- Scaling is the One of Most important step before Model Development
- Scaling means makes all the columns under one scale
- Scaling used to make all  the columns or features comparable
- Some ML models works in distance methods
  
    - example age min:0 max:100
      
    - Income might be lakhs crores so much bigger values
 
      
    - If we dont make agea and income under one scale ML model treats Income is the Important varaible
 
      
    - When values are huge maths makes more complex, so it is beter to do lower down the values
 
      
    - all the features under one scale so easy to compare
 
      
        - dollars and rupees we cannot compare because two are different scales

```python
# step 1 x=prevailing data 
# step 2 mean = mean of prevailange dat 
# step sd= sd of prevailange dat 
#step 4 step1-step2 = x-mean
#step 5 step4/step3 = {x-mean}/sd
```

```python
x=df['prevailing_wage']
u=x.mean()
sd=x.std()
z=(x-u)/sd
z
```

```python
z.hist()
```

```python
df['prevailing_wage'].hist()
```

```python
z.mean(),z.std()
```

```python
plt.subplot(1,2,1).hist(z)
plt.subplot(1,2,2).hist(df['prevailing_wage'])
plt.show()
```
